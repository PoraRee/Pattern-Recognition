&emsp; **OT1.** Note how at point B and D we have feature maps of the same size. If you think carefully, having 3x3 followed by 3x3 will cover a region of 5x5 in the input. This is why in Inception v2, the 5x5 convolution is replaced by two 3x3 convolutions. Compute the amount of multiplies to compute A and B. Compare it to the amount of multiplies to compute D. Compute the amount of parameters in the path to A and B. Compare it to the amount of parameters in path to D.

&emsp; The amount of multiplies = kernel size x output shape <br>
&emsp; to compute A: the amount of multiplies = 3x3 x 98x98x5 = 432180 multiplies <br>
&emsp; to compute B: the amount of multiplies = 3x3 x 96x96x10 = 864360 multiplies <br>
&emsp; to compute A and B: the amount of multiplies = 432180 + 864360 = 1296540 multiplies <br>
&emsp; to compute D: the amount of multiplies = 5x5 x 96x96x10 = 2304000 multiplies <br><br>

&emsp; The amount parameters = ( kernel size + 1(is a bias) ) x filters <br>
&emsp; in the path A: The amount parameters = (3x3 + 1) x 5 = 50 parameters <br>
&emsp; in the path B: The amount parameters = (3x3 + 1) x 10 = 100 parameters <br>
&emsp; in the path A and B: The amount parameters = 50 + 100 = 150 parameters <br>
&emsp; in the path D: The amount parameters = (5x5 + 1) x 10 = 260 parameters <br>
